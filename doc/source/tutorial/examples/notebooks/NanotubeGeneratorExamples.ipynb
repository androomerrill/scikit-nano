# NanotubeGenerator Examples

```python
from sknano.nanogen import NanotubeGenerator

# the following imports are only needed for displaying rich content within notebook
from IPython.display import FileLink, Image, Latex, Math
```

### For the first example, we'll generate a unit cell for a (10, 5) single-walled carbon nanotube

```python
Math(r'\mathbf{C}_{\mathrm{h}} = (10, 5)')
```

```python
nt = NanotubeGenerator(n=10, m=5)
nt.save_data(fname='data/10,5_unit_cell.xyz')
```

```python
FileLink('data/10,5_unit_cell.xyz')
```
